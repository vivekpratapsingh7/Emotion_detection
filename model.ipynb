```python
Training_data = pd.read_csv('/content/train.txt', sep=';', header=None)
Testing_data = pd.read_csv('/content/test.txt', sep=';', header=None)
validate_data = pd.read_csv('/content/val.txt', sep=';', header=None)
```

```python
Training_data.head()
```

```python
keep_labels = ['happy', 'sadness', 'anger', 'neutral']
Training_data = Training_data[Training_data[1].isin(keep_labels)]
Testing_data = Testing_data[Testing_data[1].isin(keep_labels)]
validate_data = validate_data[validate_data[1].isin(keep_labels)]
```

```python
Training_data.head()
```

```python
X_train = Training_data[0]
y_train = Training_data[1]
X_test = Testing_data[0]
y_test = Testing_data[1]
```

```python
y_train.head()
```

```python
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
```

```python
print(X_test.iloc[0])
print(y_test.iloc[0])
print(y_pred[0])
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)
```
